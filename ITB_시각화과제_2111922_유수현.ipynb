```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

mpl.rcParams['axes.unicode_minus'] = False 

path = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'  
font_name = fm.FontProperties(fname=path, size=10).get_name()
print(font_name)

plt.rc('font', family=font_name)
```

```python
fm._rebuild()
item_count = health['연령대코드(5세단위)'].value_counts()
item_count.plot.bar(rot=0)
plt.title('연령대')
plt.legend()
```

* 시각화 결과 해석: 검진자의 연령대의 경우, 11번 그룹 (50~54세)의 수가 약 1750에 가까운 수치로 가장 많았으며, 40대와 60대가 1500의 수치로 그 뒤를 이었다. 검진자 중 40대 미만의 연령대는 확인되지 않았다.

```python
fm._rebuild()
item_count = health['음주여부'].value_counts()
item_count.plot.bar(rot=0)
plt.title('음주여부')
plt.legend()
```

* 시각화 결과 해석: 전체 가입자 중에서 음주를 하는 사람의 수는 약 6000명으로, 음주를 하지 않는 사람의 수보다 약 1.5배 더 많은 것을 알 수 있다.

```python
fm._rebuild()
item_count = health['흡연상태'].value_counts()
item_count.plot.bar(rot=0)
plt.title('흡연상태')
plt.legend()
```

* 시각화 결과 해석: 전체 가입자 중에서 비흡연자의 수가 약 6000명으로 가장 많으며, 과거에 흡연을 했으나 현재 금연을 하는 가입자의 수와와 현재 흡연을 하는 가입자의 수는 약 2000명으로 비슷하다.

```python
sns.scatterplot(data = health, x = 'HDL콜레스테롤', y = 'LDL콜레스테롤').set(xlim=(20,100), ylim=(50,250))
```

* 시각화 결과 해석: HDL콜레스테롤과 LDL콜레스테롤 수치 사이에는 특별한 상관관계가 발견되지 않았다. 
HDL콜레스테롤의 정상치를 보인 사람들이 대부분이었으나, 정상치 범위를 벗어난 65~90mg/dL 수치를 보인 경우도 많았다.
반면, LDL콜레스테롤의 경우 고LDL혈증에 해당하는 170mg/dL 이상 수치를 보인 검진자 수는 많지 않았다.

```python
sns.lineplot(data = health, x = '연령대코드(5세단위)', y = '식전혈당(공복혈당)')
```

* 시각화 결과 해석: 연령대가 높아질수록 식전혈당(공복혈당) 수치가 높아지는 경향을 보인다. 그러나 80세를 기점으로 식전혈당 수치가 감소하는 것을 확인할 수 있다.

```python
sns.barplot(data = health, x = '성별코드', y = '식전혈당(공복혈당)')
```

* 시각화 결과 해석: 남성의 식전혈당(공복혈당) 수치는 약 110으로, 여성의 수치보다 미세하게 높음을 알 수 있다.

```python
fm._rebuild()
x = health['구강검진수검여부'].value_counts()
y=['No','Yes']
plt.pie(x, labels=y, autopct='%.1f%%')
plt.show
```

* 시각화 결과 해석: 검진자가 구강검진을 선택하여 검진하였는지 여부에 대한 항목 분석 결과, 미수검(0)한 비율이 수검(1)한 비율보다 약 2배 더 많은 것이 확인되었다.
